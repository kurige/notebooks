```julia
using Wavelets
using Images
using PyPlot
```

```julia
n = 2^11
x0 = testfunction(n,"Doppler")
x = x0 + 0.05*randn(n)
y = denoise(x,TI=true)
f, ax = plt.subplots(3,1, sharex=true)

cc=0.6
ax[1][:plot](x0,"b")
ax[1][:set_title]("original",fontsize=11)
ax[2][:plot](x,"g")
ax[2][:set_title]("noisy",fontsize=11)
ax[3][:plot](y,"r")
ax[3][:set_title]("denoised (TI VisuShrink)",fontsize=11)
for i=1:length(ax)
    ax[i][:set_xlim]([0,n])
    ax[i][:set_ylim]([-cc,cc])
end
f[:set_size_inches](6,5)
```

```julia
J = 11
n = 2^J
x = testfunction(n,"Bumps")
y = dwt(x, waveletls("cdf9/7"))
d,l = wplotdots(y, 0.1, n)
A = wplotim(y)

f, ax = plt.subplots(3, 1, sharex=true)
ax[1][:plot](x, "k")
ax[1][:set_xlim]([0,n])
ax[1][:set_ylabel](L"signal $x$")
ax[2][:scatter](d,l)
ax[2][:margins](0.05)
ax[2][:set_ylim](ax[2][:get_ylim]()[2:-1:1])
ax[2][:set_ylabel](L"level $j$")
ax[3][:imshow](A,aspect="auto",interpolation="none")
ax[3][:set_ylim]([J-0.5,-0.5])
ax[3][:set_ylabel](L"level $j$")

f[:set_size_inches](6,5)
```
