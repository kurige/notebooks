```julia
data = readtable("data/nasdaq.csv")

# Convert date column into an actual date object
data[:Date] = [Date(data[i, :Date]) for i=1:nrow(data)]

data
```

```julia
# Closing day prices
plot(x = data[:Date], y = data[:Close], Geom.line)
```
